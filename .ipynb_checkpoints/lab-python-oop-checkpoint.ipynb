## Challenge 1: Bank Account

Create a BankAccount class with the following attributes and methods:

Attributes:

- account_number (a unique identifier for the bank account)
- balance (the current balance of the account. By default, is 0)

Methods:

- deposit(amount) - adds the specified amount to the account balance
- withdraw(amount) - subtracts the specified amount from the account balance
- get_balance() - returns the current balance of the account
- get_account_number() - returns the account number of the account

Instructions:

- Create a BankAccount class with the above attributes and methods.
- Test the class by creating a few instances of BankAccount and making deposits and withdrawals.
- Ensure that the account_number attribute is unique for each instance of BankAccount.

*Hint: create a class attribute account_count. The account_count class attribute is used to keep track of the total number of bank accounts that have been created using the BankAccount class. Every time a new BankAccount object is created, the account_count attribute is incremented by one. This can be useful for various purposes, such as generating unique account numbers or monitoring the growth of a bank's customer base.*

```python
# ------------------------------------------
# BankAccount Class and Testing
# ------------------------------------------

class BankAccount:
    # Class attribute to track number of accounts created
    account_count = 0

    def __init__(self, balance=0):
        """Initialize a new bank account with a unique account number and optional initial balance."""
        BankAccount.account_count += 1
        self.account_number = BankAccount.account_count
        self.balance = balance

    def deposit(self, amount):
        """Add the specified amount to the account balance."""
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        """Subtract the specified amount from the account balance."""
        if amount <= 0:
            print("Withdrawal amount must be positive.")
            return
        if amount > self.balance:
            print("Insufficient funds.")
            return
        self.balance -= amount
        print(f"Withdrew ${amount:.2f}. Remaining balance: ${self.balance:.2f}")

    def get_balance(self):
        """Return the current balance of the account."""
        return self.balance

    def get_account_number(self):
        """Return the account number."""
        return self.account_number


# ------------------------------------------
# Testing the BankAccount class
# ------------------------------------------

# Create two accounts
account1 = BankAccount(1000)
account2 = BankAccount(500)

# Display initial balances
print("Account 1 balance:", account1.get_balance())
print("Account 2 balance:", account2.get_balance())

# Make deposits
account1.deposit(200)
account2.deposit(300)

# Make withdrawals
account1.withdraw(150)
account2.withdraw(50)

# Display final balances
print("\nFinal Balances:")
print(f"Account {account1.get_account_number()} balance: ${account1.get_balance():.2f}")
print(f"Account {account2.get_account_number()} balance: ${account2.get_balance():.2f}")

# Display unique account numbers
print("\nAccount Numbers:")
print("Account 1 number:", account1.get_account_number())
print("Account 2 number:", account2.get_account_number())
```

## Challenge 2: Savings Account

Create a SavingsAccount class that inherits from the BankAccount class. The SavingsAccount class should have the following additional attributes and methods:

Attributes:

- interest_rate (the annual interest rate for the savings account. By default - if no value is provided - sets it to 0.01)

Methods:

- add_interest() - adds the interest earned to the account balance
- get_interest_rate() - returns the interest rate for the account

Instructions:

- Create a SavingsAccount class that inherits from the BankAccount class and has the above attributes and methods.
- Test the class by creating a few instances of SavingsAccount and making deposits and withdrawals, as well as adding interest.

```python
# ------------------------------------------
# BankAccount Base Class
# ------------------------------------------

class BankAccount:
    # Class attribute to track number of accounts created
    account_count = 0

    def __init__(self, balance=0):
        """Initialize a new bank account with a unique account number and optional initial balance."""
        BankAccount.account_count += 1
        self.account_number = BankAccount.account_count
        self.balance = balance

    def deposit(self, amount):
        """Add the specified amount to the account balance."""
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        """Subtract the specified amount from the account balance."""
        if amount <= 0:
            print("Withdrawal amount must be positive.")
            return
        if amount > self.balance:
            print("Insufficient funds.")
            return
        self.balance -= amount
        print(f"Withdrew ${amount:.2f}. Remaining balance: ${self.balance:.2f}")

    def get_balance(self):
        """Return the current balance of the account."""
        return self.balance

    def get_account_number(self):
        """Return the account number."""
        return self.account_number


# ------------------------------------------
# SavingsAccount Subclass
# ------------------------------------------

class SavingsAccount(BankAccount):
    def __init__(self, balance=0, interest_rate=0.01):
        """Initialize a new savings account with an interest rate."""
        super().__init__(balance)  # Call the parent constructor
        self.interest_rate = interest_rate

    def add_interest(self):
        """Calculate and add interest to the account balance."""
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest of ${interest:.2f} added. New balance: ${self.balance:.2f}")

    def get_interest_rate(self):
        """Return the current interest rate."""
        return self.interest_rate


# ------------------------------------------
# Testing the SavingsAccount class
# ------------------------------------------

# Create two savings accounts with different interest rates
savings1 = SavingsAccount(1000, 0.03)  # 3% interest
savings2 = SavingsAccount(500)         # Default 1% interest

# Display initial info
print(f"Account {savings1.get_account_number()} - Balance: ${savings1.get_balance():.2f}, Interest Rate: {savings1.get_interest_rate()*100:.1f}%")
print(f"Account {savings2.get_account_number()} - Balance: ${savings2.get_balance():.2f}, Interest Rate: {savings2.get_interest_rate()*100:.1f}%")

# Perform transactions
savings1.deposit(500)
savings2.deposit(300)
savings1.withdraw(200)
savings2.withdraw(100)

# Add interest
print("\nAdding interest...")
savings1.add_interest()
savings2.add_interest()

# Final balances
print("\nFinal Balances:")
print(f"Account {savings1.get_account_number()} balance: ${savings1.get_balance():.2f}")
print(f"Account {savings2.get_account_number()} balance: ${savings2.get_balance():.2f}")
```

### Challenge 3: Checking Account

Create a CheckingAccount class that inherits from the BankAccount class. The CheckingAccount class should have the following additional attributes and methods:

Attributes:

- transaction_fee (the fee charged per transaction. By default is 1$)
- transaction_count (the number of transactions made in the current month)

Methods:

- deduct_fees() - deducts transaction fees from the account balance based on the number of transactions made in the current month and the transaction fee per transaction. The method calculates the total fees by multiplying the transaction count with the transaction fee, and deducts the fees from the account balance if it's sufficient. If the balance is insufficient, it prints an error message. Otherwise, it prints how much it's been deducted. After deducting the fees, the method resets the transaction count to 0.
- reset_transactions() - resets the transaction count to 0
- get_transaction_count() - returns the current transaction count for the account

Instructions:

- Create a CheckingAccount class that inherits from the BankAccount class and has the above attributes and methods.
- Test the class by creating a few instances of CheckingAccount and making deposits, withdrawals, and transactions to deduct fees.

Note: To ensure that the transaction count is updated every time a deposit or withdrawal is made, we need to overwrite the deposit and withdraw methods inherited from the BankAccount class. 

```python
# ------------------------------------------
# Base Class: BankAccount
# ------------------------------------------

class BankAccount:
    account_count = 0

    def __init__(self, balance=0):
        """Initialize a new bank account with unique account number."""
        BankAccount.account_count += 1
        self.account_number = BankAccount.account_count
        self.balance = balance

    def deposit(self, amount):
        """Add the specified amount to the account balance."""
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        """Subtract the specified amount from the account balance."""
        if amount <= 0:
            print("Withdrawal amount must be positive.")
            return
        if amount > self.balance:
            print("Insufficient funds.")
            return
        self.balance -= amount
        print(f"Withdrew ${amount:.2f}. Remaining balance: ${self.balance:.2f}")

    def get_balance(self):
        """Return the current balance of the account."""
        return self.balance

    def get_account_number(self):
        """Return the unique account number."""
        return self.account_number


# ------------------------------------------
# Subclass: CheckingAccount
# ------------------------------------------

class CheckingAccount(BankAccount):
    def __init__(self, balance=0, transaction_fee=1):
        """Initialize a checking account with transaction fee."""
        super().__init__(balance)
        self.transaction_fee = transaction_fee
        self.transaction_count = 0

    def deposit(self, amount):
        """Deposit money and increase transaction count."""
        super().deposit(amount)
        if amount > 0:
            self.transaction_count += 1

    def withdraw(self, amount):
        """Withdraw money and increase transaction count."""
        if amount > 0 and amount <= self.balance:
            super().withdraw(amount)
            self.transaction_count += 1
        else:
            super().withdraw(amount)  # Let parent handle errors

    def deduct_fees(self):
        """Deduct transaction fees based on the number of transactions."""
        total_fees = self.transaction_count * self.transaction_fee
        if total_fees == 0:
            print("No transactions this month. No fees deducted.")
            return
        if total_fees > self.balance:
            print("Insufficient funds to deduct transaction fees.")
            return
        self.balance -= total_fees
        print(f"Deducted ${total_fees:.2f} in transaction fees. New balance: ${self.balance:.2f}")
        self.transaction_count = 0  # Reset count after deduction

    def reset_transactions(self):
        """Reset transaction count to 0."""
        self.transaction_count = 0
        print("Transaction count has been reset to 0.")

    def get_transaction_count(self):
        """Return the current transaction count."""
        return self.transaction_count


# ------------------------------------------
# Testing the CheckingAccount class
# ------------------------------------------

# Create two checking accounts
checking1 = CheckingAccount(500)
checking2 = CheckingAccount(1000, transaction_fee=2)

# Perform deposits and withdrawals
checking1.deposit(200)
checking1.withdraw(100)
checking1.deposit(50)

checking2.withdraw(300)
checking2.deposit(400)
checking2.withdraw(200)

# Check balances and transaction counts
print("\nTransaction Summary:")
print(f"Account {checking1.get_account_number()} - Balance: ${checking1.get_balance():.2f}, Transactions: {checking1.get_transaction_count()}")
print(f"Account {checking2.get_account_number()} - Balance: ${checking2.get_balance():.2f}, Transactions: {checking2.get_transaction_count()}")

# Deduct fees for both accounts
print("\nDeducting Fees...")
checking1.deduct_fees()
checking2.deduct_fees()

# Final balances
print("\nFinal Balances After Fees:")
print(f"Account {checking1.get_account_number()} balance: ${checking1.get_balance():.2f}")
print(f"Account {checking2.get_account_number()} balance: ${checking2.get_balance():.2f}")
```

Example of testing CheckingAccount:

    - Create a new checking account with a balance of 500 dollars and a transaction fee of 2 dollars
    - Deposit 100 dollars into the account 
    - Withdraw 50 dollars from the account 
    - Deduct the transaction fees from the account

    - Deposit 200 dollars into the account
    - Withdraw 75 dollars from the account

Expected output:

    Transaction count: 0

Note: *the print "Transaction fees of 4$ have been deducted from your account balance" is done in the method deduct_fees*

```python
# ------------------------------------------
# Base Class: BankAccount
# ------------------------------------------

class BankAccount:
    account_count = 0

    def __init__(self, balance=0):
        """Initialize a new bank account with unique account number."""
        BankAccount.account_count += 1
        self.account_number = BankAccount.account_count
        self.balance = balance

    def deposit(self, amount):
        """Add the specified amount to the account balance."""
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        self.balance += amount

    def withdraw(self, amount):
        """Subtract the specified amount from the account balance."""
        if amount <= 0:
            print("Withdrawal amount must be positive.")
            return
        if amount > self.balance:
            print("Insufficient funds.")
            return
        self.balance -= amount

    def get_balance(self):
        """Return the current balance of the account."""
        return self.balance

    def get_account_number(self):
        """Return the unique account number."""
        return self.account_number


# ------------------------------------------
# Subclass: CheckingAccount
# ------------------------------------------

class CheckingAccount(BankAccount):
    def __init__(self, balance=0, transaction_fee=1):
        """Initialize a checking account with transaction fee."""
        super().__init__(balance)
        self.transaction_fee = transaction_fee
        self.transaction_count = 0

    def deposit(self, amount):
        """Deposit money and increase transaction count."""
        super().deposit(amount)
        if amount > 0:
            self.transaction_count += 1

    def withdraw(self, amount):
        """Withdraw money and increase transaction count."""
        if amount > 0 and amount <= self.balance:
            super().withdraw(amount)
            self.transaction_count += 1
        else:
            super().withdraw(amount)

    def deduct_fees(self):
        """Deduct transaction fees based on transaction count."""
        total_fees = self.transaction_count * self.transaction_fee
        if total_fees == 0:
            print("No transactions to deduct fees for.")
            return
        if total_fees > self.balance:
            print("Insufficient funds to deduct transaction fees.")
            return
        self.balance -= total_fees
        print(f"Transaction fees of {total_fees}$ have been deducted from your account balance.")
        self.transaction_count = 0

    def reset_transactions(self):
        """Reset transaction count to 0."""
        self.transaction_count = 0

    def get_transaction_count(self):
        """Return the current transaction count."""
        return self.transaction_count


# ------------------------------------------
# Example Testing of CheckingAccount
# ------------------------------------------

# Step 1: Create a new checking account with $500 balance and $2 fee
checking = CheckingAccount(500, transaction_fee=2)

# Step 2: Deposit $100
checking.deposit(100)

# Step 3: Withdraw $50
checking.withdraw(50)

# Step 4: Deduct fees
checking.deduct_fees()

# Step 5: Print current balance and transaction count
print("\nCurrent balance:", int(checking.get_balance()))
print("Transaction count:", checking.get_transaction_count())

# Step 6: Deposit $200
checking.deposit(200)

# Step 7: Withdraw $75
checking.withdraw(75)

# Step 8: Deduct fees again
checking.deduct_fees()

# Step 9: Print current balance and transaction count again
print("\nCurrent balance:", int(checking.get_balance()))
print("Transaction count:", checking.get_transaction_count())
```
